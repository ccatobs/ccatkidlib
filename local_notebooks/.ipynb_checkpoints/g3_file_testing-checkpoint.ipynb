```python
UDP_IP = "192.168.3.40"

UDP_PORT = 4096 
```

```python
sock = socket.socket(socket.AF_INET, # Internet

           socket.SOCK_DGRAM) # UDP

sock.bind((UDP_IP, UDP_PORT)) 
```

```python
def capturePacket(byteshift=-1):
    """
    Q - Remove byte rolling and conversion to separate function?
    """

    data = sock.recv(9000) # buffer size is 1024 bytes

    data = bytearray(data)

    data = np.roll(data,byteshift)

    return np.frombuffer(data, dtype="<i").astype("float")


def getNpackets(N,byteshift=-1):

    packets = capturePacket(byteshift)

    for p in range(N):

        newPacket = capturePacket(byteshift)

        packets = np.vstack([packets,newPacket])

    return packets


def getNpackets_2(N,byteshift=-1):
    
    return np.array([capturePacket(byteshift) for p in range(N)])

def livePlot():
    from matplotlib import pyplot as plt
    from IPython.display import clear_output
    import numpy as np
    plt.figure()
    for i in range(10):
        clear_output(wait=True)
        y=getNpackets_2(48)
        print (i)
        print (y.shape)
        #for j in range(1):
        #plt.plot(range(i*48,(i+1)*48),y[:, 16+1*2]**2.0+y[:, 17+1*2]**2.0)
        plt.plot(y[:, 16+1*2]**2.0+y[:, 17+1*2]**2.0)
        plt.show()
```

```python
frames = []

reader = so3g.G3IndexedReader('test_output.g3')

while True:

    frame = reader.Process(None)
    assert len(frame) <= 1
    if len(frame) == 0:
        break
    frames.extend(frame)
```

The file size is now 509326603 bytes, or 486 MB...clearly, this is going to get too large very quickly unless we do some repackaging. We would get a GB every three seconds otherwise, which is a TB every 51 minutes. We might also be able to store the data more efficiently than casing every packet to a G3VectorDouble as an array. We also might be able to use compression and downsampling here - it would be fairly easy to average two packets together and cut the data in half by 2 while only dropping the sampling rate to 256 Hz, which is still above the roughly 200 Hz that SO is using.

The next thing that I want to try is streaming directly into a g3 file where the packets are added in as they are received.

Would there be less overhead to store each value in the packet as a separate G3Double instead of wrapping the whole packet as a G3VectorDouble? I am not sure how time-limited we are here, so it isn't clear to me yet how much processing we can do before we start missing packets. I'm also not sure how to know if we do miss a packet. Are they labeled coming out of the RFSoC? Adrian is working on adding timestamping, which would help.

```python
def stream_to_g3(N_samples,byteshift=-1):
    writer = core.G3Writer('test_stream.g3')

    # Metadata frame
    frame = core.G3Frame(core.G3FrameType.Housekeeping)
    frame['label'] = core.G3String("Test output file for RFSoC packets")
    writer(frame)

    for i in range(N_samples):
        packet = capturePacket(byteshift) # gets packet from socket
        frame = core.G3Frame(core.G3FrameType.Housekeeping)
        frame['name'] = core.G3String("Packet number {}".format(i))
        frame['data'] = core.G3VectorDouble(packet) # stores packet as G3VectorDouble
        writer(frame) # Writes frame to g3 file

    writer(core.G3Frame(core.G3FrameType.EndProcessing))
    writer = None
```
