```python
from ccatkidlib.analysis.vna import VNA
from ccatkidlib.analysis.target import Target
from ccatkidlib.analysis.timestream import Timestream
from pathlib import Path
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
root_dir = Path('/home/rfsoc/ccatkidlib_data/loopback')
data_dir = 'loopback_rfsoc102'
date = '20250528'
sess_id = '1748461305'

bid = '1'
drid='3'
```

```python
vna_path = root_dir / data_dir / date / sess_id / 'vna' / f'B{bid}D{drid}'
```

```python
plt.figure()
for file in vna_path.iterdir():
    vna = VNA(com_to = f'{bid}.{drid}', data_path = [file])
    if vna.drone_cfg['atten']['sense'] == 14:
        plt.plot(vna.fs(), 20*np.log10(vna.mag()))
```
